```python
import os
import shutil
import random
```

```python
def mix_images(source_dir, destination_dir):
    # Get all subfolders (representing classes) in the source directory
    class_folders = os.listdir(destination_dir)
    
    for class_folder in class_folders:
        print(class_folder)
        source_class_path = os.path.join(source_dir, class_folder)
        destination_class_path = os.path.join(destination_dir, class_folder)
        
        # Check if source is a directory and destination class folder exists
        if os.path.isdir(source_class_path):
            # Create the destination class folder if it doesn't exist
            if not os.path.exists(destination_class_path):
                print(f"{destination_class_path} doesn't exist.")
                os.makedirs(destination_class_path)
            
            # Get all images in the source class folder
            image_files = os.listdir(source_class_path)
            print(f'Number of images: {len(image_files)}')
            
            i = 0
            for image_file in image_files:
                source_image_path = os.path.join(source_class_path, image_file)
                destination_image_path = os.path.join(destination_class_path, image_file)
                
                # Check if it's a file (not a directory)
                if os.path.isfile(source_image_path):
                    # Copy the image to the destination
                    shutil.copy2(source_image_path, destination_image_path)
                    i += 1
            print(f'{i} images were copied into {destination_class_path}.\n')
    
    print("Image mixing completed.")
```

```python
# Example usage:
source_dir = os.path.join('..', 'datasets', 'splitted', 'train')
destination_dir = os.path.join('..', 'datasets', 'extended_dataset', 'train')

mix_images(source_dir, destination_dir)
```

```python
def copy_test_folder(source_dir, destination_dir):
    try:
        shutil.copytree(source_dir, destination_dir)
        print(f'Folder successfully copied from {source_dir} into {destination_dir}')
    except Exception as e:
        print(f'Error {e}')
```

```python
target_source_dir = os.path.join('..', 'datasets', 'splitted', 'test')
destination_source_dir = os.path.join('..', 'datasets', 'extended_dataset', 'test_classified')

copy_test_folder(target_source_dir, destination_source_dir)
```
